```python
import numpy as np
from matplotlib import pyplot as plt

def sig(x):
  return 1 / (1 + np.exp(-x))


def val(x, w, b):
    """
    input:
        `N`[1...inf) input rows
    return:
        `1D` arr of `N` predictions
    """
    return np.dot(w, x.T) + b  # w is applied to N[1...inf) rows of x. Result: 1D arr of N predictions

def pred(x,w,b):
  return sig(val(x,w,b))
dataset = np.array([
  [0,0],
  [1,0],
  [2,0],
  [3,1],
  [4,1],
  [5,1],
])
x_train = dataset[:,:-1]
y_train = dataset[:,-1:]
# Depending on w and b, the step in the plot will look different
w = 100
b = -250
preds = pred(np.arange(0,10,0.01), w, b)
plt.plot(np.arange(0,10,0.01),preds)
plt.grid()
plt.show()
```
