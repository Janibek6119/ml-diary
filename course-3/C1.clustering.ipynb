```python
import numpy as np
from matplotlib import pyplot as plt
from C1 import create_2D_clusters, calculate_cost

np.random.seed(1337)
a, b = create_2D_clusters(5, 100)
c = np.vstack((a, b))
K = np.random.rand(2, 2)
K_history = [K]
J_history = []
plt.scatter(a[:,0],a[:,1])
plt.scatter(b[:,0],b[:,1], c="red")
plt.show()
```

```python
cost, examples_0, examples_1 = calculate_cost(c, K)

MIDS = np.array([np.mean(examples_0, axis=0), np.mean(examples_1, axis=0)])

plt.scatter(examples_0[:, 0], examples_0[:, 1])
plt.scatter(examples_1[:, 0], examples_1[:, 1], c="purple")
plt.scatter(K[0, 0], K[0, 1], marker='X', c="green")
plt.scatter(K[1, 0], K[1, 1], marker='X', c="red")
plt.scatter(MIDS[0, 0], MIDS[0, 1], marker='x', c="green")
plt.scatter(MIDS[1, 0], MIDS[1, 1], marker='x', c="red")
plt.show()
K=MIDS
J_history.append(cost)
K_history.append(K)
```

```python
K_history_arr = np.array(K_history)
K_history_delta = K_history_arr[1:] - K_history_arr[:-1]
K_history_delta_magnitude = np.sqrt(np.sum(np.array(K_history_delta)**2, axis=2))
plt.plot(K_history_delta_magnitude[:,0], c="green")
plt.plot(K_history_delta_magnitude[:,1], c="red")
plt.grid()
plt.show()
plt.plot(J_history, c="purple")
plt.grid()
plt.show()
```
